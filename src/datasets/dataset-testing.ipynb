# Finding the best height and width

```python
import numpy as np
import matplotlib.pyplot as plt
import math

from FGVC_aircraft_dataset import AircraftCaptionDataset

dataset = AircraftCaptionDataset("C:\\CodeProjects\\CS-4644-Project\\src\\datasets\\fgvc-aircraft-dataset\\")

max_H: int = -math.inf
max_W: int = -math.inf
for i in range(10000):
    if i % 100 == 0:
        print(i)
    
    datapoint = dataset.__getitem__(i)
    H, W, C = datapoint[0].shape
    if H > max_H:
        max_H = H
    if W > max_W:
        max_W = W

print(f"max_H: {max_H}\nmax_W: {max_W}")
```
